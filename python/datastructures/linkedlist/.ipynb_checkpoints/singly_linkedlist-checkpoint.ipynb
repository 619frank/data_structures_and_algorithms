```python
import json


class Node:
    def __init__(self, value):
        self.value = value
        self.next = None


class SinglyLinkedList:

    def __init__(self):
        self.head = None
        self.tail = None
        self.length = 0

    
    def push(self, value):
        newNode = Node(value)
        if self.head is None:
            self.head = newNode
            self.tail = newNode            
        else:
            self.tail.next = newNode
            self.tail = newNode
        
        self.length += 1
        
    def pop(self):
        if self.head is None:
            return
        current = self.head
        newTail = current
        
        while current.next:
            newTail = current
            current = current.next
        
        self.tail = newTail
        newTail.next = None
        self.length -= 1
        return current
    
    def shift(self):
        if self.head is None:
            return False
        
        self.head = self.head.next
        
        self.length -= self.length
    
    def unshift(self,value):
        
        if self.head is None:
            self.push(value)
            return True
        
        newHead = Node(value)
        newHead.next = self.head
        self.head = newHead        
        self.length += self.length
        
        return True
    
    def get(self, index):
        if self.length == 0:
            return False
        if self.length == 1:
            return self.head.index
        
        current = self.head
        
        for x in range(index):
            current = current.next
            
        return current.value
    
    def set(self, index, value):
        if self.length == 0:
            self.push(value)
            return True
        current = self.head
        for x in range(index):
            current= current.next
        current.value = value
        return True
        
        

sll = SinglyLinkedList()

sll.push(1)

sll.push(2)

sll.push(3)
sll.push(4)

# print(sll.shift())
sll.unshift(10)
print(sll.get(2))
# print(sll.get(2))
# print(json.dumps(sll, default=lambda obj: vars(obj),  indent=2))
```
